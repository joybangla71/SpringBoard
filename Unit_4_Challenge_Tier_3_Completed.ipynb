You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

Let's first confirm the borough names are indedd the indexes.

```python
properties_transposed.index
```

Yes we are right the borough names are the indexes. We can do `reset_index()` on the transposed DataFrame to solve this issue.

```python
properties_transposed = properties_transposed.reset_index()
```

```python
properties_transposed.index
```

Great! Our `properties_transposed` DataFrame indecies are no longer our borough names i.e., the indicies are now a numerical RangeIndex. Now lets peek at the DataFrame using head to confirm that resetting index worked.

```python
properties_transposed.head(10)
```

It worked but looking at the `propeties_transposed` DataFrame it is obvious that the columns are also integer values whereas we want the months to be our column names. Lets see the current column names first and then we will fix that.

```python
properties_transposed.columns
```

Yep! Our column names are not what we want at the moment. Let us replace the current column names with content from row 0 in our DatFrame as this row contains the months.

```python
properties_transposed.columns = properties_transposed.iloc[0]
properties_transposed.head()
```

Excellent! Now we see that row 0 is repeated as the column names and also the first row of the `properties_transposed` DataFrame. Let's drop the first row then.

```python
properties_transposed = properties_transposed.drop(0)
properties_transposed.head()
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great!

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
df_yearly_mean_price[(df_yearly_mean_price['London_Borough'] == 'Croydon') & (df_yearly_mean_price['Year'] == 1998)]['Average_price'].tolist()[0]
```

```python
df_yearly_mean_price[(df_yearly_mean_price['London_Borough'] == 'Croydon') & (df_yearly_mean_price['Year'] == 2018)]['Average_price'].tolist()[0]
```

```python
370298.54332 / 88143.22883
```

```python
def get_price_ratio(df, borough_name, year_start, year_end):
  return df[(df['London_Borough'] == borough_name) & (df['Year'] == year_end)]['Average_price'].tolist()[0] / df[(df['London_Borough'] == borough_name) & (df['Year'] == year_start)]['Average_price'].tolist()[0]
```

Test out the function:

```python
get_price_ratio(df_yearly_mean_price, 'Croydon', 1998, 2018)
```

```python
get_price_ratio(df_yearly_mean_price, 'Camden', 1998, 2018)
```

Now we will use our function to compute the average prices separated by 20 years. Let us select 2022 and 2002:

```python
last_year = 2022
twenty_years_ago = 2002
data_dict = {}

for b in df_yearly_mean_price['London_Borough'].unique():
  data_dict[b] = get_price_ratio(df_yearly_mean_price, b, twenty_years_ago, last_year)

print(data_dict)
```

Let us create a dataframe from the `data_dict` dictionary so that we can see the ratios better.

```python
ratio_df = pd.DataFrame(list(data_dict.items()))
ratio_df
```

Looks good but notice that the columns names are 0 and 1. Let's change these:

```python
ratio_df.rename(columns={0: "Borough", 1: 'Ratio at 2022'}, inplace=True)
ratio_df.head()
```

Let's now plot out data.

```python
ax = ratio_df[['Borough', 'Ratio at 2022']].plot(kind='bar')
ax.set_xticklabels(ratio_df.Borough)
```
